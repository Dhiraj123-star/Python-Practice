```python
import pandas as pd

# first dataframe
d = {

    'Employee_id':[1,2,3,4,5],
    'Employee_name':['Akshar','Jones','Kate','Mike','Meera']
}
df1 = pd.DataFrame(d,columns=['Employee_id','Employee_name'])

df1
```

```python

# second dataframe

data = {

    'Employee_id':[4,5,6,7,8],
    'Employee_name':['Meera','Tia','Varsha','Williams','Ziva']

}

df2 = pd.DataFrame(data,columns=['Employee_id','Employee_name'])

df2
```

```python
# merging 

pd.merge(df1,df2,on="Employee_id")
```

```python
# grouping 


data = {
    'Employee_id':[1,2,3,4,5],
    'Employee_name':["Meera","Akash","Rohit","Roshan","Meera"]
} 
df2 = pd.DataFrame(data)
group = df2.groupby('Employee_name') 
group.get_group('Meera')
```

```python
# concatenating 
pd.concat([df1,df2])
```

```python
zoo_eats = pd.read_csv("zoo_eats.csv",delimiter=";")

zoo_eats
```

```python
# merge two dataframes

zoo.merge(zoo_eats)
```

```python
zoo_eats.merge(zoo) # by default inner join
```

```python
zoo.merge(zoo_eats,how='outer') # outer join
```

```python
zoo.merge(zoo_eats,how='left') # left join
```

```python
zoo.merge(zoo_eats,how='right')
```

```python
zoo.merge(zoo_eats,how='left',left_on='animal',right_on='animal') # recognise(left_on,right_one) the key-columns from the datasets
```

```python
# sorting 

zoo.sort_values('water_need') # by default in ascending order
```

```python
# sort with multiple columns 

zoo.sort_values(by=['animal','water_need'])
```

```python
zoo.sort_values(by=['water_need'],ascending=False) # sort in descending order
```

```python
# reset index

zoo.sort_values(by=['water_need'],ascending=False).reset_index()
```

```python
zoo.sort_values(by=['water_need'],ascending=False).reset_index(drop=True) # drop the index column
```
