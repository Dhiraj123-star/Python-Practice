```python
# melt function

# Create a simple dataframe

# importing pandas as pd
import pandas as pd

# creating a dataframe
df1 = pd.DataFrame({'Name': {0: 'John', 1: 'Bob', 2: 'Shiela'},
				'Course': {0: 'Masters', 1: 'Graduate', 2: 'Graduate'},
				'Age': {0: 27, 1: 23, 2: 21}})
df1
```

```python
pd.melt(df1,id_vars=['Name'],value_vars=['Course'])
```

```python
pd.melt(df1,id_vars=['Course'],value_vars=['Age'])
```

```python
# name of 'variable' and 'value' columns can be customised

pd.melt(df1,id_vars=['Name'],value_vars=['Course'],var_name='ChangedVarName',value_name='ChangedValueName')
```
