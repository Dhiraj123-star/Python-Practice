```python
# for else

# traditional approach 

numbers = [1,2,3,4]

allEven = True

for number in numbers :
    if number%2==1:
        allEven=False
        print('No')
        break
if allEven:
    print('Even')
```

```python
# modern approach (with for-else) 

numbers = [0,2,4,6]

for number in numbers:
    if number%2==1:
        print('No')
        break
else:
    print('Yes')
```

```python
# yield 
# generate millions of generate numbers

def generate_ids(limit):
    for id in range(1,limit+1):
        yield id

ids = generate_ids(10000000)

# this is easily managed with yield keyword 
print(next(ids))
print(next(ids))
print(next(ids))
print(next(ids))
print(next(ids))
print(next(ids))
print(next(ids))
print(next(ids))
```

```python
# comparision chaining
'''way to link multiple comparision statements without using and keyword'''

a,b,c =1,2,3
d,e,f =5,5,3

print(a<b<c)  # true
print(a==b==c) # false
print(a<d<=e) # true
print(d==e!=f) # true
print(d>f<e) # true 
```

```python
# Timeit 

''' 
Timeit executes the code N number of times and gives a better value for the runtime.
By default, the code is executed 1 Million times

'''

import timeit

Code = ''' 
def fun():
    sum = 0 
    for i in range(100000000000):
        sum+=i
    
    return sum

'''

print(f'The execution time {timeit.timeit(stmt=Code)}')
```

```python
# decorators 
'''
It is used to modifying the behaviour of a function or class without changing the code.

'''

import time

def runtime_decorator(func):
    def wrapper(*args,**kwargs):
        start = time.time()
        ret =func(*args)
        end = time.time()

        print(f'The function {func.__name__} took {end-start} seconds to execute. ')
        return ret
    return wrapper

@runtime_decorator
def generateNumbers(n):
    return [i for i in range(n)]

generateNumbers(1000000)
```

```python
# zfill 

# it is used to pad 0's to the left of a string to make a string of a specific length

import datetime

today = datetime.datetime.now()
day =today.day
month = today.month
year = today.year

print(f"{day}-{month}-{year}")
print(f'{str(day).zfill(2)}-{str(month).zfill(2)}-{str(year).zfill(2)}') # add 0 as prefix
```

```python
# any/all
# any behaves like of series of OR conditions and all behaves like AND conditions

numbers=[23,45,67,100]

# check if any of the value in numbers is odd 

print(any([i%2==1 for i in numbers])) # True

# check if all of the value in numbers is odd

print(all([i%2==1 for i in numbers])) # False
```
